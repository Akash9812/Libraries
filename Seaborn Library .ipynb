```python
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
df.head()
```

### Correlations and Heatmap 
A correlation heatmap uses colored cells. typically in a monochromatic scale to show a 2D correlation matrix(table) between two discrete dimensions or events types. It is very important in feature selections.

```python
df.corr()
```

```python
df.dtypes
```

```python
sns.heatmap(df.corr())
```

### Jointplot 
A joint plot allows to study the relationship between 2 numeric variables. The central chart display their correlation. It is usually a scatterplot, a 2D Histogram, a density plot 

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg')
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')
```

## Pair plot 
A pair plot is also known as sactter plot, in which one variable in the same data row is matched with another variable's value, like this: pairs plots are just elaborations on this, showing all variables paired with all the other variables

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='day')
```

## Dist Plot
Dist plot helps to check the distribtion of the columns feature

```python
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'],kde=False,bins=10)
```

## Categorical Plots 
Seaborn also helps us in doing the analysis on categorical data points in this section we will discuss about 
* Boxplot
* countplot
* bar plot
* Violinplot

```python
# Countplot
sns.countplot('sex',data=df)
```

```python
sns.countplot(y='sex',data=df)
```

```python
#Barplot
import numpy as np
sns.barplot(x='day',y='total_bill',data=df)
```

```python
A=df.total_bill[df.day=='Thur'].std()/np.sqrt(len(df.total_bill[df.day=='Thur']))
B=df.total_bill[df.day=='Thur'].mean()
print(A,B)
```

```python

print(B+2*A)
print(B-2*A)
```

### Box Plot
A boxplot represents the information from a five-number summary 

```python
sns.boxplot('smoker','total_bill',data=df)
```

```python
sns.boxplot(x='day',y='total_bill',data=df,palette='rainbow')
```

```python
# categorize data based on some other categories
sns.boxplot(x='total_bill',y='day',hue='smoker',data=df)
```

## Violin Plot
Violin plot helps us to see both the distribution of data in terms of karnel density estimation and the box plot

```python
sns.violinplot(x='total_bill',y='day',data=df,palette='rainbow')
```
